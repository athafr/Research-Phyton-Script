# 9. K-Means Clustering

9.1 - Import Modul K-Means

```python
from sklearn.cluster import KMeans
from sklearn import metrics
```

9.2 - Import Data

```python
data_c = pd.read_excel('D:/Kuliah ITS/Kuliah/Semester 8/Thesis/Prog Thesis/Data Saham ISSI/Next/Trend Cluster/Trend Clustering Result.xlsx')
data_c
```

```python
# Pilih column yang digunakan dalam clustering saja
x = data_c.iloc[:,1:9]
x
```

9.3 - Standarisasi Data

```python
# Import modul preprocessing
from sklearn import preprocessing
```

```python
x_scaled = preprocessing.scale(x)
x_scaled
```

9.3 - Clustering

```python
kmeans = KMeans(11)
```

```python
kmeans.fit(x_scaled)
```

9.4 - Menentukan Jumlah Klaster Yang Optimal

9.4.1 - WCSS Method

```python
kmeans.inertia_
```

```python
wcss = []
cl_num = 16
for i in range (2,cl_num):
    kmeans= KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
    
wcss
```

```python
#Visualisasi
number_clusters = range(2,cl_num)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

9.4.2 Silhoutte Method

```python
metrics.silhouette_score(x_scaled, kmeans.labels_, metric='euclidean')
```

```python
silhouette = []
cl_num = 16
for i in range (2,cl_num):
    kmeans= KMeans(i)
    kmeans.fit(x_scaled)
    silhouette_iter = metrics.silhouette_score(x_scaled, kmeans.labels_, metric='euclidean')
    silhouette.append(silhouette_iter)
    
silhouette
```

```python
silhouette = pd.DataFrame(silhouette)
silhouette
```

```python
#Visualisasi
number_clusters = range(2,cl_num)
plt.plot(number_clusters, silhouette)
plt.title('The Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
```

```python
# Silhouette visualizer
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
cl_num = 11
for i in range (2,cl_num):
    kmeans= KMeans(i)
    visualizer = SilhouetteVisualizer(kmeans)
    visualizer.fit(x_scaled)
    visualizer.poof()
```

```python
kmeans= KMeans(6)
visualizer = SilhouetteVisualizer(kmeans)
visualizer.fit(x_scaled)
visualizer.poof()
```

9.5 - Cluster Result

```python
identified_clusters = kmeans.fit_predict(x_scaled)
identified_clusters
```

```python
data_with_clusters = data_c.copy()
data_with_clusters['Cluster'] = identified_clusters
data_with_clusters
```

Visualize the result

```python
plt.scatter(data_with_clusters['All Quarterly Risk Std.'], data_with_clusters['All Quarterly Return'], c=data_with_clusters['Cluster'], cmap = 'rainbow')
```

```python
# Visualisasi Heatmap
import seaborn as sns

sns.clustermap(data_c, cmap='mako')
```

# 10. Portofolio Return & Risk

```python
# Portfolio cov matrix
pfolio_daily_returns = pd.read_excel('D:/Kuliah ITS/Kuliah/Semester 8/Thesis/Prog Thesis/Data Saham ISSI/Next/Newest Portfolio/Daily Return Portofolio.xlsx', index_col = 'Date')
pfolio_daily_returns
```

```python
pfolio_daily_returns.info()
```

```python
pfolio_cov_matrix = pfolio_daily_returns.cov() * 55
pfolio_cov_matrix
```

```python
pfolio_corr_matrix = pfolio_daily_returns.corr()
pfolio_corr_matrix
```

```python
pfolio_quarterly_returns = pfolio_daily_returns.mean() * 55
pfolio_quarterly_returns
```

```python
weights = np.array([0.25, 0.25, 0.25, 0.25])
```

```python
weights[0]
```

```python
#Portfolio Return
pfolio_ret = np.dot(pfolio_quarterly_returns, weights)
pfolio_ret
```

```python
#Portfolio Volatility
pfolio_vol = (np.dot(weights.T, np.dot(pfolio_daily_returns.cov() * 55, weights))) ** 0.5
pfolio_vol
```

```python
#Portfolio Variance
pfolio_var = np.dot(weights.T, np.dot(pfolio_daily_returns.cov() * 55, weights))
pfolio_var
```

# Portofolio Diversible & Non-Diversible Risk

Prepare Variance of each stock

```python
a_var = pfolio_daily_returns['HRUM.JK'].var() * 55
a_var
```

```python
b_var = pfolio_daily_returns['MDKA.JK'].var() * 55
b_var
```

```python
c_var = pfolio_daily_returns['PYFA.JK'].var() * 55
c_var
```

```python
d_var = pfolio_daily_returns['EMTK.JK'].var() * 55
d_var
```

Calculate diversified risk

```python
#Diversible Risk
dr = pfolio_var - (weights[0] ** 2 * a_var) - (weights[1] ** 2 * b_var) - (weights[2] ** 2 * c_var) - (weights[3] ** 2 * d_var)
dr
```

```python
#Non-Diverisble Risk
n_dr = pfolio_var - dr
n_dr
```
